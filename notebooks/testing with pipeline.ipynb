```python
import joblib

model = joblib.load("fake_news_model.pkl")
vectorizer = joblib.load("tfidf_vectorizer.pkl")
encoder = joblib.load("label_encoder.pkl")
```

```python
def predict_news(text):

    text_vector = vectorizer.transform([text])

    pred = model.predict(text_vector)[0]

    label = encoder.inverse_transform([pred])[0]

    return label
```

```python
news = """
The government announced a new healthcare policy today.
"""

print(
    predict_news(news)
)
```

```python
news = """
Scientists discovered dragons living on Mars.
"""

print(
    predict_news(news)
)
```

```python
def predict_news_with_confidence(text):

    text_vector = vectorizer.transform([text])

    probs = model.predict_proba(text_vector)[0]

    pred_idx = probs.argmax()

    label = encoder.inverse_transform([pred_idx])[0]

    confidence = probs.max()

    print("Prediction :", label)
    print("Confidence :", round(confidence*100,2), "%")

    print("\nClass Probabilities")

    for cls, prob in zip(encoder.classes_, probs):
        print(f"{cls}: {prob:.4f}")
```

```python
predict_news_with_confidence(
    "Scientists discovered dragons living on Mars."
)
```

```python
def predict_news_with_threshold(text):

    text_vector = vectorizer.transform([text])

    probs = model.predict_proba(text_vector)[0]

    pred_idx = probs.argmax()

    confidence = probs.max()

    label = encoder.inverse_transform([pred_idx])[0]

    if confidence < 0.60:
        return "Uncertain", confidence

    return label, confidence
```

```python
predict_news_with_threshold(
    "Dragon on Mars"
)
```
